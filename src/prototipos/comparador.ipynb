```python
PATH = "../datos/"
TEST_FILE = "exoTest.csv"
df_test = pd.read_csv(f'{PATH}{TEST_FILE}', low_memory=False)
df_test['LABEL'] = df_test['LABEL'].add(-1)
fourier_sizes = [1598, 1000, 250, 20]
df_test.iloc[:,1:] = df_test.iloc[:,1:].apply(z_score_normalizing, axis = 1)
df_fourier = df_test.iloc[:,:(df_test.shape[1]//2)]
```

```python
models_folder = model_name = os.getcwd() + "\\saved_models\\"
for file in os.listdir(models_folder):
    model_name = models_folder + file[:-4]
    print()
    print(file[:-4])
    if "lstm" in file:
        modelo = ModeloLSTM()
        test_model_lstm(modelo, model_name, df_test)
    elif "_fourier" in file:
        modelo = Perceptron(fourier_sizes)
        test_model(modelo, model_name, df_fourier)
    elif "_bce" in file:
        modelo = Perceptron(output = 1)
        test_model(modelo, model_name, df_test)
    else:
        modelo = Perceptron()
        test_model(modelo, model_name, df_test)
    print("******************************************************************************************************")
```
